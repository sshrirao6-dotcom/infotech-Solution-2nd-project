```python
import pandas as pd
import numpy as np

# Load clean price data
prices = pd.read_csv('clean_stock_prices.csv', index_col=0, parse_dates=True)

prices.head()
```

```python
# Calculate Daily Returns
# (How much did the price change compared to yesterday?)
daily_returns = prices.pct_change()

daily_returns.head()
```

```python
# Cleaning
daily_returns = daily_returns.dropna()
```

```python
daily_returns.head()
```

```python
# Calculate Volatility (Risk)
volatility = daily_returns.std()

volatility
```

```python
volatility_percent = volatility * 100
volatility_percent
```

```python
# Monte Carlo Simulation
stock = 'AAPL'
# stock = 'MSFT'
# stock = 'BTC-USD'
```

```python
mean_return = daily_returns[stock].mean()
std_return = daily_returns[stock].std()

mean_return, std_return
```

```python
num_simulations = 1000
num_days = 252  # 1 trading year

simulated_prices = []

last_price = prices[stock].iloc[-1]

for i in range(num_simulations):
    daily_simulated_returns = np.random.normal(mean_return, std_return, num_days)
    price_path = last_price * (1 + daily_simulated_returns).cumprod()
    simulated_prices.append(price_path)
```

```python
mc_results = []

for sim_no, path in enumerate(simulated_prices):
    for day, value in enumerate(path):
        mc_results.append([sim_no + 1, day + 1, value])

mc_df = pd.DataFrame(
    mc_results,
    columns=["Simulation", "Day", "Portfolio_Value"]
)
```

```python
mc_df
```

```python
# Calculate Value at Risk (VaR)

# Calculate final prices:(extracts the last (final-day) price)
final_prices = [path[-1] for path in simulated_prices]
```

```python
# Calculate VaR:
var_95 = np.percentile(final_prices, 5)

print("Value at Risk (95%):", var_95)
```

```python
import pandas as pd

var_df = pd.DataFrame({
    "Confidence_Level": ["95%"],
    "VaR_Value": [var_95]
})
var_df.to_csv("var_summary.csv", index=False)
```

```python
# Recalculate rolling volatility properly
rolling_vol = daily_returns.rolling(window=30).std()

# Convert index (Date) into a column
rolling_vol = rolling_vol.reset_index()

# Convert from wide to long format (BEST for Tableau)
rolling_vol_long = rolling_vol.melt(
    id_vars="Date",
    var_name="Stock",
    value_name="Rolling_Volatility"
)

# Remove initial NaN rows (first 29 days)
rolling_vol_long = rolling_vol_long.dropna()

# Save clean file
rolling_vol_long.to_csv(
    "rolling_volatility.csv",
    index=False,
    encoding="utf-8"
)

```
